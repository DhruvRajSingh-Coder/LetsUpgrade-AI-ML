```
import pandas as pd
Data=pd.read_csv("https://raw.githubusercontent.com/LetsUpgrade/AI-ML-July-2020/master/Day-7/Assignment/general_data.csv")
print(Data)
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Data["Attrition"]=le.fit_transform(Data["Attrition"])
Data.drop('EmployeeID',axis=1,inplace=True)
Data.head(4410)
```

```
 print(Data.corr() ["Attrition"])
```

___From the above correlation coefficients we can conclude that:___
'Age', 'BusinessTravel', 'Department', 'DistanceFromHome', 'Education', 'JobLevel', 'MonthlyIncome', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager' are negatively correlated with 'Attrition' and 'NumCompaniesWorked', 'PercentSalaryHike' are positively correlated with 'Attrition'.Further 'Employeecount' and 'StandardHours' are not correlated with 'Attrition'.
